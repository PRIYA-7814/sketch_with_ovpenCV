```python
import cv2
from google.colab.patches import cv2_imshow

# Sketch function
def sketch(image):
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    img_gray_blur = cv2.GaussianBlur(img_gray, (5, 5), 0)
    canny_edges = cv2.Canny(img_gray_blur, 50, 150)
    ret, mask = cv2.threshold(canny_edges, 100, 255, cv2.THRESH_BINARY_INV)
    return mask

# Load image (make sure 'flowers.jpg' is uploaded)
image = cv2.imread('flowers.jpg')
if image is None:
    print("Image not found or not loaded properly.")
else:
    result = sketch(image)
    cv2_imshow(result)
```
